1. Urči pořadí jednotlivých kandidátů v jednotlivých státech a v jednotlivých letech (pomocí metody `rank()`). Nezapomeň, že data je před použitím metody nutné seřadit a spolu s metodou `rank()` je nutné použít metodu `groupby()`

```python

df["poradi"] = df.groupby(["state", "year"])["candidatevotes"].rank(method="min", ascending=False)
```

Pro další analýzu jsou důležití pouze vítězové. Vytvoř novou tabulku, která bude obsahovat pouze vítěze voleb.
(Dale jsem pridala jeste ze zadani "důležité následující sloupce") - tabulka s nazvem vitezove

```python

new_df = df[["year", "state", "party_simplified", "candidatevotes", "totalvotes", "poradi"]]
vitezove = new_df[new_df["poradi"] == 1]
vitezove.head(20)
```

```python
vitezove = vitezove.sort_values(["state", "year"])
vitezove.head(20)
```

Pomocí metody `shift()` přidej nový sloupec, abys v jednotlivých řádcích měl(a) po sobě vítězné strany ve dvou po sobě jdoucích letech.

```python
vitezove["minule vyhrála strana"] = vitezove.groupby("state")["party_simplified"].shift()
vitezove
```

Porovnej, jestli se ve dvou po sobě jdoucích letech změnila vítězná strana. Můžeš k tomu použít např. funkci `numpy.where()` nebo metodu `apply()`.

```python
def shoda(vitezove):
    if vitezove["party_simplified"] == vitezove["minule vyhrála strana"]:
        return 0
    else:
        return 1
```

```python
vitezove = vitezove.dropna() #muzu si dovolit vyhodit ty prvni (rok 1076) protoze uz je mam v "minule vyhrala strana" (?)
vitezove["zmena"] = vitezove.apply(shoda, axis=1)
vitezove
```

Proveď agregaci podle názvu státu a seřaď státy podle počtu změn vítězných stran


```python
vitezove["zmeny_cumsum"] = vitezove.groupby("state")["zmena"].cumsum()
pocet_zmen = vitezove[vitezove["year"] == 2020]
new_tab = pocet_zmen[["state", "zmeny_cumsum"]]
new_tab.sort_values("zmeny_cumsum", ascending=False, ignore_index=True)
```

Vytvoř sloupcový graf s 10 státy, kde došlo k nejčastější změně vítězné strany. Jako výšku sloupce nastav počet změn

- proč s 10???

```python
graf_data = new_tab.sort_values("zmeny_cumsum", ascending=False).set_index("state").head(10)

graf_data["zmeny_cumsum"].plot.bar()
```

#### DRUHÁ ČÁST ÚKOLU

Pro další část pracuj s tabulkou se dvěma nejúspěšnějšími kandidáty pro každý rok a stát (tj. s tabulkou, která oproti té minulé neobsahuje jen vítěze, ale i druhého v pořadí).

1. Přidej do tabulky sloupec, který obsahuje absolutní rozdíl mezi vítězem a druhým v pořadí

```python
new_df = df[["year", "state", "party_simplified", "candidatevotes", "totalvotes", "poradi"]]
data_druha_cast = new_df[(new_df["poradi"] == 1)|(new_df["poradi"] == 2)]
data_druha_cast["candidatevotes2"] = data_druha_cast.groupby("state")["candidatevotes"].shift()

data_druha_cast["absolutni rozdil"] = abs(data_druha_cast["candidatevotes"] - data_druha_cast["candidatevotes2"])

data_druha_cast = data_druha_cast[["year", "state", "party_simplified", "totalvotes", "absolutni rozdil"]]
data_druha_cast = data_druha_cast.dropna()
data_druha_cast
```

Přidej sloupec s relativním marginem, tj. rozdílem vyděleným počtem hlasů

```python
data_druha_cast["relativní margin"] = data_druha_cast["absolutni rozdil"]/data_druha_cast["totalvotes"]
data_druha_cast
```

Seřaď tabulku podle velikosti relativního marginu a zjisti, kdy a ve kterém státě byl výsledek voleb nejtěsnější. 
 --> čím menší margin tím víc těsné (doufam?) - tzn Florida roku 2000

```python
data_druha_cast.sort_values(by="relativní margin", ascending = True)
```

Vytvoř pivot tabulku, která zobrazí pro jednotlivé volební roky, kolik států přešlo od Republikánské strany k Demokratické straně, kolik států přešlo od Demokratické strany k Republikánské straně a kolik států volilo kandidáta stejné strany.

```python
vitezove
```

```python
def swing(vitezove):
    if vitezove["zmena"] == 1:
        if vitezove["party_simplified"] == "REPUBLICAN":
            return "to REP"
        else:
            return "to DEM"
    else:
        return "no swing"
```

```python
vitezove["swing"] = vitezove.apply(swing, axis = 1)
vitezove
```

```python
pivot_tabulka = pandas.pivot_table(data=vitezove, index="year", values="poradi", columns="swing", aggfunc=sum, fill_value=0)
pivot_tabulka
```
